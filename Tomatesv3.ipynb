```python
import os

# Configuration basée sur ton YAML
# Remplace 'TON_DOSSIER' par le nom du dossier créé par le !curl
BASE_PATH = "/content/TON_DOSSIER"
INDEX_FAUTE = 0  # 'Anthrocnose'
INDEX_VRAI = 3   # 'anthracnose'

def fix_labels(folder):
    label_path = os.path.join(BASE_PATH, folder, 'labels')
    if not os.path.exists(label_path):
        return

    count = 0
    for filename in os.listdir(label_path):
        if filename.endswith(".txt"):
            file_path = os.path.join(label_path, filename)
            with open(file_path, 'r') as f:
                lines = f.readlines()

            new_lines = []
            for line in lines:
                parts = line.split()
                if parts[0] == str(INDEX_FAUTE):
                    parts[0] = str(INDEX_VRAI)
                    count += 1
                new_lines.append(" ".join(parts) + "\n")

            with open(file_path, 'w') as f:
                f.writelines(new_lines)
    print(f"[{folder}] Terminé : {count} étiquettes corrigées.")

# Exécution sur les 3 dossiers
for split in ['train', 'valid', 'test']:
    fix_labels(split)
```

```python
import os

# Configuration pour ton projet à la racine de Colab
BASE_PATH = "/content"
INDEX_FAUTE = "0"  # 'Anthrocnose' (avec la faute)
INDEX_VRAI = "3"   # 'anthracnose' (le bon)

def merge_tomato_classes():
    splits = ['train', 'valid', 'test']
    total_corrected = 0

    for split in splits:
        label_path = os.path.join(BASE_PATH, split, 'labels')

        if not os.path.exists(label_path):
            print(f"⚠️ Dossier non trouvé : {label_path}")
            continue

        print(f"🔄 Traitement du dossier : {split}...")

        for filename in os.listdir(label_path):
            if filename.endswith(".txt"):
                file_path = os.path.join(label_path, filename)

                # Lecture du fichier
                with open(file_path, 'r') as f:
                    lines = f.readlines()

                new_lines = []
                modified = False

                for line in lines:
                    parts = line.split()
                    if len(parts) > 0 and parts[0] == INDEX_FAUTE:
                        parts[0] = INDEX_VRAI
                        new_lines.append(" ".join(parts) + "\n")
                        modified = True
                        total_corrected += 1
                    else:
                        new_lines.append(line)

                # Réécriture seulement si une modification a été faite
                if modified:
                    with open(file_path, 'w') as f:
                        f.writelines(new_lines)

    print(f"\n✅ Terminé ! {total_corrected} étiquettes ont été fusionnées vers l'index {INDEX_VRAI}.")

# Lancement de l'opération
merge_tomato_classes()
```
